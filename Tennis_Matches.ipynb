```python
#checking size of df
df = pd.concat([pd.read_csv(f) for f in files])
print(df.shape)
```

```python
#making a separate year column
dfs = []
for f in files:
    temp = pd.read_csv(f)
    temp["year"] = f.split("_")[2].replace(".csv", "")  # adjust based on your filename pattern
    dfs.append(temp)

df = pd.concat(dfs, ignore_index=True)
```

```python
df['year'] = df['year'].astype(int)
```

```python
#make sure all data types are valid
df.dtypes
```

```python
df["w_ace"].apply(type).value_counts()
```

```python
#Filter out for only masters and grand slam level tournaments
mastersPlus = df[
    (df['tourney_level'] == 'M') |
    (df['tourney_level'] == 'G')
    ].sort_values(['tourney_date', 'match_num']).reset_index(drop=True)

mastersPlus
```

```python
#Drop all unnecessary columns
mastersPlus = mastersPlus.drop(columns = ['tourney_id', 'draw_size', 'tourney_level', 'tourney_date', 'winner_id', 'winner_entry', 'winner_hand', 'loser_id', 'loser_entry', 'loser_hand', 'winner_rank', 'winner_rank_points', 'loser_rank', 'loser_rank_points'])

mastersPlus
```

```python
from google.colab import files
#Export df into csv file
pd.DataFrame(mastersPlus).to_csv('mastersPlus.csv', index=False)

#Download file to computer rather than colab storage
files.download('mastersPlus.csv')
```

```python
#Filter out only for Round of 16(end goal is to get only the top 8 players -> winners of R16)
R16Winners = mastersPlus[
    (mastersPlus['round'] == 'R16')
]

R16Winners
```

```python
print(R16Winners[['tourney_name','year','winner_name']])
```

```python
#Get all winners of R16 into a set per tournament
Top8Winners = R16Winners.groupby(['tourney_name', 'year'])['winner_name'].apply(set).reset_index()

Top8Winners.head(50)
```

```python
Top8Winners.columns = ['tourney_name', 'year', 'top8_players']

Top8Winners
```

```python
#Put the set of players back into the original df
dfMerged = pd.merge(mastersPlus, Top8Winners, on=['tourney_name', 'year'], how='left')

dfMerged
```

```python
#Run through each tournament and only keep rows that match with the top 8 players from the tournament
Filtered = dfMerged[dfMerged.apply(lambda x: x['winner_name'] in x['top8_players'], axis=1)].drop(columns=['top8_players'])

Filtered
```

```python
from google.colab import files
#Export df into csv file
pd.DataFrame(Filtered).to_csv('Top8_TourneyData.csv', index=False)

#Download file to computer rather than colab storage
files.download('Top8_TourneyData.csv')
```

```python
#Drop a bunch of other unnecessary cols - also adding more, so no need for extraneous info
Filtered = Filtered.drop(columns = ['match_num', 'winner_seed', 'loser_seed', 'best_of', 'minutes', 'winner_ioc', 'loser_ioc'])
#Add new statistical columns:
#Total points won throughout match
Filtered['winner_total_points'] = (
    Filtered['w_1stWon'] + Filtered['w_2ndWon'] + (Filtered['l_svpt'] - Filtered['l_1stWon'] - Filtered['l_2ndWon'])
)
Filtered['loser_total_points'] = (
    Filtered['l_1stWon'] + Filtered['l_2ndWon'] + (Filtered['w_svpt'] - Filtered['w_1stWon'] - Filtered['w_2ndWon'])
)
#Points difference - higher numbers is better/more dominant
Filtered['points_difference'] = (
    Filtered['winner_total_points'] - Filtered['loser_total_points']
)
#1st serve percentages - % of total serves that landed in first time
Filtered['w_1stSv%'] = (Filtered['w_1stIn'] / Filtered['w_svpt'] * 100).round(2)
Filtered['l_1stSv%'] = (Filtered['l_1stIn'] / Filtered['l_svpt'] * 100).round(2)
#1st serve percentage difference - higher number is better/more first time accuracy
Filtered['1stSv%_difference'] = (Filtered['w_1stSv%'] - Filtered['l_1stSv%']).round(2)
#1st serve win percentage - %of first time serves won by the server
Filtered['w_1stSvWin%'] = (Filtered['w_1stWon'] / Filtered['w_1stIn'] * 100).round(2)
Filtered['l_1stSvWin%'] = (Filtered['l_1stWon'] / Filtered['l_1stIn'] * 100).round(2)
#Difference equation - higher number is better/more dominant
Filtered['1stSvWin%_difference'] = (Filtered['w_1stSvWin%'] - Filtered['l_1stSvWin%']).round(2)
#2nd serve win percentage - %of 2nd time serves won by the server
Filtered['w_2ndSvWin%'] = ((Filtered['w_2ndWon']) / (Filtered['w_svpt'] - Filtered['w_1stIn']) * 100).round(2)
Filtered['l_2ndSvWin%'] = ((Filtered['l_2ndWon']) / (Filtered['l_svpt'] - Filtered['l_1stIn']) * 100).round(2)
#Difference equation - higher number is better/More secure with 2nd serves
Filtered['2ndSvWin%_difference'] = (Filtered['w_2ndSvWin%'] - Filtered['l_2ndSvWin%']).round(2)
#% of aces served by winner - higher is better
Filtered['ace%'] = (Filtered['w_ace'] / (Filtered['w_ace'] + Filtered['l_ace']) * 100).round(2)
#% of break points faced - lower is better, more secure with service game
Filtered['bpFaced%'] = (Filtered['w_bpFaced'] / (Filtered['w_bpFaced'] + Filtered['l_bpFaced']) * 100).round(2)
```

```python
stat_cols = [
    'w_ace', 'w_df', 'w_svpt', 'w_1stIn', 'w_1stWon', 'w_2ndWon',
    'w_bpSaved', 'w_bpFaced', 'winner_total_points', 'w_1stSv%', 'w_1stSvWin%',
    'w_2ndSvWin%', 'w_SvGms',
    'l_ace', 'l_df', 'l_svpt', 'l_1stIn', 'l_1stWon', 'l_2ndWon',
    'l_bpSaved', 'l_bpFaced', 'loser_total_points', 'l_1stSv%', 'l_1stSvWin%',
    'l_2ndSvWin%', 'l_SvGms',
    'points_difference', '1stSv%_difference', '1stSvWin%_difference',
    '2ndSvWin%_difference', 'ace%', 'bpFaced%'
]
Top8_Aggregated = Filtered.groupby(['tourney_name', 'year'])[stat_cols].mean().round(2)

#Deltas are first calculated by row-level and then are recalculated when aggregating
Top8_Aggregated['ace%'] = (Top8_Aggregated['w_ace'] / (Top8_Aggregated['w_ace'] + Top8_Aggregated['l_ace']) * 100).round(2)
Top8_Aggregated['bpFaced%'] = (Top8_Aggregated['w_bpFaced'] / (Top8_Aggregated['w_bpFaced'] + Top8_Aggregated['l_bpFaced']) * 100).round(2)
Top8_Aggregated['points_difference'] = (Top8_Aggregated['winner_total_points'] - Top8_Aggregated['loser_total_points'])
Top8_Aggregated['1stSv%_difference'] = (Top8_Aggregated['w_1stSv%'] - Top8_Aggregated['l_1stSv%']).round(2)
Top8_Aggregated['1stSvWin%_difference'] = (Top8_Aggregated['w_1stSvWin%'] - Top8_Aggregated['l_1stSvWin%']).round(2)
Top8_Aggregated['2ndSvWin%_difference'] = (Top8_Aggregated['w_2ndSvWin%'] - Top8_Aggregated['l_2ndSvWin%']).round(2)
Top8_Aggregated
```
